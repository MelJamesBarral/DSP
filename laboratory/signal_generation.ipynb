```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11)

delta = np.zeros_like(n)
delta[n == 0] = 1

plt.stem(n, delta, use_line_collection=True)
plt.title('Unit Sample Sequence (Delta Function)')
plt.xlabel('n')
plt.ylabel('Delta[n]')
plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11, 1)

u = np.zeros_like(n)
u[n >= 0] = 1

plt.plot(n, u, drawstyle='steps-post', color='blue', linewidth=2)
plt.title('Unit Step Signal (u[n])')
plt.xlabel('n')
plt.ylabel('u[n]')
plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.5)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11, 1)

r = np.zeros_like(n)
r[n >= 0] = n[n >= 0]

plt.plot(n, r, color='blue', linewidth=2)
plt.title('Unit Ramp Signal (r[n])')
plt.xlabel('n')
plt.ylabel('r[n]')
plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-1, 10)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(-2, 2, 1000)
a = 1
exponential_signal = a * np.exp(t)


plt.figure(figsize=(10, 5))
plt.plot(t, exponential_signal, label='Exponential Signal', color='r')
plt.title('Exponential Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.ylim(-1, 8)
plt.xlim(-2, 2)
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square


t = np.linspace(0, 2, 1000)
frequency = 1
square_wave = square(2 * np.pi * frequency * t)


plt.figure(figsize=(10, 5))
plt.plot(t, square_wave, label='Square Wave Signal', color='b')
plt.title('Square Wave Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.ylim(-1.5, 1.5)
plt.xlim(0, 2)
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth


t = np.linspace(0, 2, 1000)
frequency = 1
triangular_wave = sawtooth(2 * np.pi * frequency * t, 0.5)


plt.figure(figsize=(10, 5))
plt.plot(t, triangular_wave, label='Triangular Wave Signal', color='m')
plt.title('Triangular Wave Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.ylim(-1.5, 1.5)
plt.xlim(0, 2)
plt.grid()
plt.legend()
plt.show()
```
